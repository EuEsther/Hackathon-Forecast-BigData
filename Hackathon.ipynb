## 1**CRIAÇÃO E TREINO DO MODELO**

```python
# Instalar o CatBoost, se necessário
!pip install catboost

# ================================
# Importações
# ================================
import pandas as pd
import numpy as np
from catboost import CatBoostRegressor
import warnings
import datetime

warnings.filterwarnings('ignore')

# ====================================
# 1️⃣ Preparação dos dados
# ====================================
print("1️⃣  Preparando os dados para o CatBoost...")

tabela_2['transaction_date'] = pd.to_datetime(tabela_2['transaction_date'])
tabela_2['year'] = tabela_2['transaction_date'].dt.isocalendar().year
tabela_2['week'] = tabela_2['transaction_date'].dt.isocalendar().week
tabela_2['month'] = tabela_2['transaction_date'].dt.month

# Agregação semanal
weekly_sales = (
    tabela_2.groupby(['internal_store_id', 'internal_product_id', 'year', 'week', 'month'], as_index=False)
    .agg({'quantity': 'sum'})
)
weekly_sales.rename(columns={'internal_store_id': 'pdv', 'internal_product_id': 'produto'}, inplace=True)

# Merge com metadados
df = weekly_sales.merge(tabela_1, on='pdv', how='left').merge(tabela_3, on='produto', how='left')

# Colunas categóricas
categorical_cols = ['descricao', 'categoria', 'tipos', 'label', 'subcategoria', 'marca',
                    'fabricante', 'categoria_pdv', 'premise', 'zipcode']

for c in categorical_cols:
    if c in df.columns:
        df[c] = df[c].astype(str).fillna('nan')

df = df.sort_values(['pdv', 'produto', 'year', 'week']).reset_index(drop=True)

# ====================================
# 2️⃣ Criação de lags e rolling
# ====================================
print("2️⃣  Criando features de lag e rolling mean...")

for lag in [1, 2, 3, 4]:
    df[f'lag_{lag}'] = df.groupby(['pdv', 'produto'])['quantity'].shift(lag)

df['rolling_4'] = df.groupby(['pdv', 'produto'])['quantity'].shift(1).rolling(window=4, min_periods=1).mean()

df.fillna(0, inplace=True)

# ====================================
# 3️⃣ Treinar o modelo CatBoost
# ====================================
print("3️⃣  Treinando o modelo CatBoost...")

# Features e alvo
features = [c for c in df.columns if c not in ['pdv', 'produto', 'quantity', 'year', 'week']]
target = 'quantity'

# Categóricas presentes de fato
final_cat_features = [c for c in categorical_cols if c in features]

# Inicializa o modelo
model = CatBoostRegressor(
    iterations=500,
    learning_rate=0.05,
    depth=10,
    loss_function='RMSE',
    verbose=100,
    random_seed=42,
    allow_writing_files=False
)

# Treinamento
model.fit(
    df[features], df[target],
    cat_features=final_cat_features
)

# ====================================
# 4️⃣ Previsão recursiva com CatBoost
# ====================================
print("4️⃣  Iniciando o processo de previsão recursiva...")

# Função para adicionar semanas à data ISO e retornar year, week, month
def add_weeks(year, week, add):
    # ISO week: Monday is the first day (1)
    first_day = datetime.datetime.strptime(f'{year}-W{int(week)}-1', "%G-W%V-%u")
    new_date = first_day + pd.Timedelta(weeks=add)
    return new_date.isocalendar().year, new_date.isocalendar().week, new_date.month

# Preparar o forecast_cart
pdv_prod = df[['pdv', 'produto']].drop_duplicates()
forecast_weeks = [1, 2, 3, 4, 5]

# Criar forecast_cart com colunas básicas
forecast_cart = pd.DataFrame(
    [(w, p, pdv) for w in forecast_weeks for pdv, p in zip(pdv_prod['pdv'], pdv_prod['produto'])],
    columns=['semana', 'produto', 'pdv']
)

# Pegar última data (year, week, month) para cada pdv-produto
last_dates = df.groupby(['pdv', 'produto'])[['year', 'week', 'month']].last().reset_index()

# Merge para adicionar essas datas
forecast_cart = forecast_cart.merge(last_dates, on=['pdv', 'produto'], how='left')

# Ajustar year, week, month para as semanas futuras
forecast_cart[['year', 'week', 'month']] = forecast_cart.apply(
    lambda row: add_weeks(row['year'], row['week'], row['semana']),
    axis=1, result_type='expand'
)

# Últimas lags e rolling
last_stats = df.groupby(['pdv', 'produto'])[['lag_1', 'lag_2', 'lag_3', 'lag_4', 'rolling_4']].last().reset_index()
forecast_cart = forecast_cart.merge(last_stats, on=['pdv', 'produto'], how='left')

# Fallbacks
mean_prod = df.groupby('produto')['quantity'].mean().to_dict()
mean_cat = df.groupby('categoria')['quantity'].mean().to_dict()
prod_to_cat = df.set_index('produto')['categoria'].to_dict()

# Adiciona features categóricas
for c in final_cat_features:
    key_col = 'produto' if c in tabela_3.columns else 'pdv'
    mapping = df.groupby(key_col)[c].last().to_dict()
    forecast_cart[c] = forecast_cart[key_col].map(mapping).astype(str).fillna('nan')

# Função de atualização dos lags
def update_lags(row, pred):
    row['lag_4'] = row['lag_3']
    row['lag_3'] = row['lag_2']
    row['lag_2'] = row['lag_1']
    row['lag_1'] = pred
    row['rolling_4'] = np.mean([row['lag_1'], row['lag_2'], row['lag_3'], row['lag_4']])
    return row

# Loop de previsão recursiva
predictions = []
for week in forecast_weeks:
    subset = forecast_cart[forecast_cart['semana'] == week].copy()

    # Fallbacks para lags ausentes
    for col in ['lag_1', 'lag_2', 'lag_3', 'lag_4', 'rolling_4']:
        subset[col] = subset[col].fillna(subset['produto'].map(mean_prod))
        subset[col] = subset[col].fillna(subset['produto'].map(lambda p: mean_cat.get(prod_to_cat.get(p, 'nan'), 0)))
    subset.fillna(0, inplace=True)

    # Previsão
    subset['quantity'] = model.predict(subset[features])
    subset['quantity'] = subset['quantity'].clip(0).round().astype(int)

    predictions.append(subset[['semana', 'pdv', 'produto', 'quantity']])

    # Atualizar lags no forecast_cart para próxima iteração
    if week < max(forecast_weeks):
        updated_rows = subset.apply(lambda row: update_lags(row, row['quantity']), axis=1)
        forecast_cart.update(updated_rows)

# Concatenar previsões
submission = pd.concat(predictions).reset_index(drop=True)

# ====================================
# 5️⃣ Exportar CSV final
# ====================================
print("5️⃣  Salvando o arquivo de submissão final...")

submission.to_csv("submission_catboost.csv", sep=';', index=False, encoding='utf-8')
print("\n✅ Processo com CatBoost concluído! CSV final gerado: submission_catboost.csv")
```

```python
# Instalar dependências
!pip install catboost lightgbm xgboost optuna

import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

from sklearn.model_selection import TimeSeriesSplit
from sklearn.metrics import mean_squared_error, mean_absolute_error
import optuna

from catboost import CatBoostRegressor
import lightgbm as lgb
import xgboost as xgb

# ================================================
# 1. Carregar dados (editar caminhos conforme necessário)
# ================================================
# Exemplo:
# tabela_1 = pd.read_csv("tabela1.csv")
# tabela_2 = pd.read_csv("tabela2.csv")
# tabela_3 = pd.read_csv("tabela3.csv")

# Exemplo de estrutura já informada:
# tabela_1 colunas: pdv, premise, categoria_pdv, zipcode
# tabela_2 colunas: internal_store_id, internal_product_id, distributor_id, transaction_date, reference_date, quantity, gross_value, net_value, gross_profit, discount, taxes
# tabela_3 colunas: produto, categoria, descricao, tipos, label, subcategoria, marca, fabricante

# ================================================
# 2. Função de preparação de dados + engenharia de features
# ================================================
def prepare_data(t1, t2, t3):
    # Renomear para consistência
    tabela1 = t1.copy()
    tabela2 = t2.copy()
    tabela3 = t3.copy()

    # Data
    tabela2['transaction_date'] = pd.to_datetime(tabela2['transaction_date'])
    tabela2['year'] = tabela2['transaction_date'].dt.isocalendar().year
    tabela2['week'] = tabela2['transaction_date'].dt.isocalendar().week
    tabela2['month'] = tabela2['transaction_date'].dt.month

    # Agregação semanal de quantidade
    weekly_sales = (
        tabela2.groupby(['internal_store_id', 'internal_product_id', 'year', 'week', 'month'], as_index=False)
        .agg({'quantity': 'sum'})
    )
    weekly_sales.rename(columns={'internal_store_id': 'pdv', 'internal_product_id': 'produto'}, inplace=True)

    # Merge com metadados
    df = weekly_sales.merge(tabela1, on='pdv', how='left').merge(tabela3, on='produto', how='left')

    # Definir colunas categóricas que você possui
    categorical_cols = ['descricao', 'categoria', 'tipos', 'label', 'subcategoria',
                        'marca', 'fabricante', 'categoria_pdv', 'premise', 'zipcode']
    for c in categorical_cols:
        if c in df.columns:
            df[c] = df[c].astype(str).fillna('nan')

    # Ordenar
    df = df.sort_values(['pdv', 'produto', 'year', 'week']).reset_index(drop=True)

    # Novas features
    # Binarizar se houve venda
    df['sale_binary'] = (df['quantity'] > 0).astype(int)

    # Frequência nas últimas 4 semanas de existência de venda
    df['freq_4w'] = (
        df.groupby(['pdv', 'produto'])['sale_binary']
          .shift(1)  # look back
          .rolling(window=4, min_periods=1)
          .sum()
          .fillna(0)
    )

    # Recência: semanas desde última venda
    # Para cada grupo, podemos calcular o número de semanas desde a venda mais recente
    def calc_weeks_since_last_sale(s):
        # s é a série sale_binary por grupo pdv-produto
        # vamos criar uma contagem decrescente (volta no tempo)
        last = None
        res = []
        for val in s:
            if val == 1:
                last = 0
            else:
                if last is None:
                    # ainda não teve venda anterior
                    last = np.nan
                else:
                    last = last + 1
            res.append(last)
        return pd.Series(res, index=s.index)

    df['weeks_since_last_sale'] = (
        df.groupby(['pdv', 'produto'])['sale_binary']
          .apply(calc_weeks_since_last_sale)
          .fillna(0)
    )

    # Lags e rolling de quantidade
    for lag in [1,2,3,4]:
        df[f'lag_{lag}'] = df.groupby(['pdv','produto'])['quantity'].shift(lag)
    df['rolling_4'] = (
        df.groupby(['pdv','produto'])['quantity']
          .shift(1)
          .rolling(window=4, min_periods=1)
          .mean()
    )

    # Preencher NaNs
    df.fillna(0, inplace=True)

    return df

# ================================================
# 3. Funções de avaliação
# ================================================
def evaluate(y_true, y_pred):
    rmse = mean_squared_error(y_true, y_pred, squared=False)
    mae = mean_absolute_error(y_true, y_pred)
    return {'rmse': rmse, 'mae': mae}

# ================================================
# 4. Otimização de hiperparâmetros para CatBoost com validação temporal
# ================================================
def optimize_catboost(df, features, target, cat_features, n_trials=20, n_splits=3):
    def objective(trial):
        params = {
            'iterations': trial.suggest_int('iterations', 200, 1000),
            'learning_rate': trial.suggest_float('learning_rate', 0.01, 0.1, log=True),
            'depth': trial.suggest_int('depth', 4, 10),
            'subsample': trial.suggest_float('subsample', 0.5, 1.0),
            'colsample_bylevel': trial.suggest_float('colsample_bylevel', 0.5, 1.0),
            'l2_leaf_reg': trial.suggest_float('l2_leaf_reg', 1e-1, 10.0, log=True),
            'random_seed': 42,
            'loss_function': 'RMSE',
            'verbose': 0
        }

        tscv = TimeSeriesSplit(n_splits=n_splits)
        rmses = []
        for train_idx, val_idx in tscv.split(df):
            X_train = df.iloc[train_idx][features]
            y_train = df.iloc[train_idx][target]
            X_val = df.iloc[val_idx][features]
            y_val = df.iloc[val_idx][target]

            model = CatBoostRegressor(**params)
            model.fit(X_train, y_train, cat_features=cat_features,
                      eval_set=(X_val, y_val), early_stopping_rounds=50, verbose=False)

            preds = model.predict(X_val)
            rmse_fold = mean_squared_error(y_val, preds, squared=False)
            rmses.append(rmse_fold)

        return np.mean(rmses)

    study = optuna.create_study(direction='minimize')
    study.optimize(objective, n_trials=n_trials)

    print("🔍 Melhor conjunto de hiperparâmetros CatBoost:", study.best_params)
    print("🔍 Melhor RMSE médio:", study.best_value)
    return study.best_params

# ================================================
# 5. Treinar múltiplos modelos para comparar
# ================================================
def train_models(df, features, target, cat_features, params_cb):
    models = {}
    metrics = {}

    # CatBoost com os melhores parâmetros
    model_cb = CatBoostRegressor(**params_cb, verbose=0)
    model_cb.fit(df[features], df[target], cat_features=cat_features)

    metrics['CatBoost'] = evaluate(df[target], model_cb.predict(df[features]))
    models['CatBoost'] = model_cb

    # LightGBM
    model_lgb = lgb.LGBMRegressor(
        n_estimators=params_cb.get('iterations', 500),
        learning_rate=params_cb.get('learning_rate', 0.05),
        max_depth=params_cb.get('depth', -1),
        subsample=params_cb.get('subsample', 1.0),
        colsample_bytree=params_cb.get('colsample_bylevel', 1.0),
        random_state=42
    )
    model_lgb.fit(df[features], df[target])
    metrics['LightGBM'] = evaluate(df[target], model_lgb.predict(df[features]))
    models['LightGBM'] = model_lgb

    # XGBoost
    model_xgb = xgb.XGBRegressor(
        n_estimators=params_cb.get('iterations', 500),
        learning_rate=params_cb.get('learning_rate', 0.05),
        max_depth=params_cb.get('depth', 6),  # XGBoost depth limitação diferente
        subsample=params_cb.get('subsample', 1.0),
        colsample_bytree=params_cb.get('colsample_bylevel', 1.0),
        random_state=42,
        verbosity=0
    )
    model_xgb.fit(df[features], df[target])
    metrics['XGBoost'] = evaluate(df[target], model_xgb.predict(df[features]))
    models['XGBoost'] = model_xgb

    return models, metrics

# ================================================
# 6. Previsão recursiva para submissão
# ================================================
def recursive_forecast(model, df, features, cat_features, t3, forecast_weeks=[1,2,3,4,5]):
    # t3 é tabela_3 para mapear categorias / etc se for necessário
    pdv_prod = df[['pdv','produto']].drop_duplicates()
    forecast_cart = pd.DataFrame(
        [(w, produto, pdv) for w in forecast_weeks for pdv, produto in zip(pdv_prod['pdv'], pdv_prod['produto'])],
        columns=['semana','produto','pdv']
    )

    # Get the last year and week from the historical data
    last_year = df['year'].max()
    last_week = df[df['year'] == last_year]['week'].max()

    # Calculate the year and week for each forecast week
    forecast_cart['year'] = last_year + (forecast_cart['semana'] + last_week -1) // 52
    forecast_cart['week'] = (forecast_cart['semana'] + last_week -1) % 52 + 1
    forecast_cart['month'] = pd.to_datetime(forecast_cart['year'].astype(str) + '-' + forecast_cart['week'].astype(str) + '-1', format='%Y-%W-%w').dt.month


    # últimos lags
    last_stats = df.groupby(['pdv','produto'])[['lag_1','lag_2','lag_3','lag_4','rolling_4', 'freq_4w', 'weeks_since_last_sale']].last().reset_index()
    forecast_cart = forecast_cart.merge(last_stats, on=['pdv','produto'], how='left')

    mean_prod = df.groupby('produto')['quantity'].mean().to_dict()
    mean_cat = df.groupby('categoria')['quantity'].mean().to_dict()
    prod_to_cat = df.set_index('produto')['categoria'].to_dict()

    # Mapear variáveis categóricas
    for c in cat_features:
        # Determinar se c é da tabela de produto ou pdv
        if c in t3.columns:
            key_col = 'produto'
        elif c in df.columns:
            key_col = 'pdv'
        else:
            key_col = None

        if key_col:
            mapping = df.groupby(key_col)[c].last().to_dict()
            forecast_cart[c] = forecast_cart[key_col].map(mapping).astype(str).fillna('nan')
        else:
            # se feature não encontrada, encher padrão
            forecast_cart[c] = 'nan'

    # Função de atualizar lags
    def update_lags(row, pred):
        row['lag_4'] = row['lag_3']
        row['lag_3'] = row['lag_2']
        row['lag_2'] = row['lag_1']
        row['lag_1'] = pred
        row['rolling_4'] = np.mean([row['lag_1'], row['lag_2'], row['lag_3'], row['lag_4']])

        # Update freq_4w and weeks_since_last_sale
        row['sale_binary'] = (pred > 0).astype(int)
        row['freq_4w'] = row['freq_4w'] - (row['lag_4'] > 0).astype(int) + row['sale_binary']
        row['weeks_since_last_sale'] = 0 if row['sale_binary'] == 1 else row['weeks_since_last_sale'] + 1

        return row

    predictions = []
    for week in forecast_weeks:
        subset = forecast_cart[forecast_cart['semana'] == week].copy()

        # Fallbacks
        for col in ['lag_1','lag_2','lag_3','lag_4','rolling_4', 'freq_4w', 'weeks_since_last_sale']:
            subset[col] = subset[col].fillna(subset['produto'].map(mean_prod))
            subset[col] = subset[col].fillna(subset['produto'].map(lambda p: mean_cat.get(prod_to_cat.get(p, 'nan'), 0)))
        subset.fillna(0, inplace=True)

        # Predição
        subset['quantity'] = model.predict(subset[features])
        subset['quantity'] = subset['quantity'].clip(0).round().astype(int)

        predictions.append(subset[['semana','pdv','produto','quantity']])

        # Atualizar lags para próximas semanas no forecast_cart
        if week < max(forecast_weeks):
            mask = forecast_cart['semana'] == week
            updated_cols = ['lag_1','lag_2','lag_3','lag_4','rolling_4', 'freq_4w', 'weeks_since_last_sale']
            updated = subset.apply(lambda row: update_lags(row, row['quantity']), axis=1)
            forecast_cart.loc[mask, updated_cols] = updated[updated_cols].values

    submission = pd.concat(predictions).reset_index(drop=True)
    return submission

# ================================================
# 7. Execução completa
# ================================================
if __name__ == "__main__":
    # Carregar dados
    # adaptar essas linhas conforme seus caminhos
    # tabela_1 = pd.read_csv("tabela1.csv")
    # tabela_2 = pd.read_csv("tabela2.csv")
    # tabela_3 = pd.read_csv("tabela3.csv")

    df = prepare_data(tabela_1, tabela_2, tabela_3)

    target = 'quantity'
    # definir features automaticamente
    non_features = ['pdv','produto','quantity','year','week']
    features = [c for c in df.columns if c not in non_features]
    categorical_cols = ['descricao', 'categoria', 'tipos', 'label', 'subcategoria',
                        'marca', 'fabricante', 'categoria_pdv', 'premise', 'zipcode']
    cat_features = [c for c in categorical_cols if c in features]

    # 1) Otimização CatBoost
    best_params_cb = optimize_catboost(df, features, target, cat_features, n_trials=20, n_splits=3)

    # 2) Treinar e comparar modelos
    models_dict, metrics = train_models(df, features, target, cat_features, best_params_cb)
    print("Desempenho dos modelos no treino completo:")
    for m, met in metrics.items():
        print(f"  {m} => RMSE: {met['rmse']:.4f}, MAE: {met['mae']:.4f}")

    # 3) Escolher modelo final (provavelmente CatBoost, se for o melhor)
    modelo_final = models_dict['CatBoost']

    # 4) Fazer previsão para próximas semanas
    submission_df = recursive_forecast(modelo_final, df, features, cat_features, tabela_3,
                                       forecast_weeks=[1,2,3,4,5])

    # 5) Salvar resultados e modelo
    submission_df.to_csv("submission_final.csv", sep=';', index=False, encoding='utf-8')
    print("✅ CSV de submissão salvo como submission_final.csv")

    modelo_final.save_model("modelo_final_catboost.cbm")
    print("✅ Modelo salvo como modelo_final_catboost.cbm")
```
